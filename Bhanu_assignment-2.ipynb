```python
import os
import glob
```

```python
#config
config_dictionary ={"Image_set":"c:\\Users\\sande\\OneDrive\\Desktop\\Data_mining\\Images",
                    "bread_info":{"n02085936-Maltese_dog":0,"n02096294-Australian_terrier":1,
                                  "n02106550-Rottweiler":2,"n02092339-Weimaraner":3
                                 }
                 }

Knn_cluster_size = {
                        1:{'cross_validation_method':{'validation_error':'','training':''},'stratified_cross_validation_method':{'validation_error':'','training':''}},
                        3:{'cross_validation_method':{'validation_error':'','training':''},'stratified_cross_validation_method':{'validation_error':'','training':''}},
                        5:{'cross_validation_method':{'validation_error':'','training':''},'stratified_cross_validation_method':{'validation_error':'','training':''}},
                        6:{'cross_validation_method':{'validation_error':'','training':''},'stratified_cross_validation_method':{'validation_error':'','training':''}},
                        10:{'cross_validation_method':{'validation_error':'','training':''},'stratified_cross_validation_method':{'validation_error':'','training':''}},
                        20:{'cross_validation_method':{'validation_error':'','training':''},'stratified_cross_validation_method':{'validation_error':'','training':''}}
                   }
```

```python
#creating a master data frame to carry the values
import pandas as pd
master_data_frame = pd.DataFrame()
```

```python
import cv2 as cv
import sys
```

```python
#loading each file and appending the label information
for bread in config_dictionary["bread_info"].keys():
    image_loading_path = str(os.path.join(config_dictionary["Image_set"],bread,"cropped_images"))
    if os.path.exists(image_loading_path):
        for file_path in glob.glob(image_loading_path+"/*.jpg"):
            loaded_image = cv.imread(file_path)
            gray_converted_image = cv.cvtColor(loaded_image, cv.COLOR_BGR2GRAY)
            image_histogram_values = cv.calcHist([gray_converted_image],[0],None,[256],[0,256])
            t = pd.DataFrame(image_histogram_values).T
            t["bread_flag"] = config_dictionary["bread_info"][bread]
            master_data_frame = master_data_frame.append(t)
    else:
        print("image Loading path is not available")
        sys.exit(1)
```

```python
master_data_frame
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_obj = StandardScaler()
df = master_data_frame.loc[:,:255]
labels = master_data_frame['bread_flag'].tolist()
df = pd.DataFrame(scaler_obj.fit_transform(df), columns=df.columns) 
df["bread_flag"] = labels
```

```python
master_data_frame = df
master_data_frame
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore",category=FutureWarning)
```

```python
#creating the global train, test, train_label and test_label instances
train = pd.DataFrame()
test = pd.DataFrame()

train_label = pd.Series()
test_label = pd.Series()
```

```python
from sklearn.model_selection import train_test_split
```

```python
bread_labels = config_dictionary["bread_info"].keys()
for bread_label in bread_labels:
    filtered_master_df = master_data_frame.query("bread_flag == {}".format(\
                                                                           config_dictionary["bread_info"][bread_label]\
                                                                          ))
    filtered_master_df_features = filtered_master_df.loc[:,:255]
    filter_master_df_labels = filtered_master_df["bread_flag"]
    
    split_train,split_test,split_train_label,split_test_label = train_test_split(filtered_master_df_features,filter_master_df_labels,test_size = 0.20)
    
    #pushing the information into global instance
    if len(train) == 0 or len(test) == 0  or len(train_label) ==0 or len(test_label) == 0 :
        train= split_train
        test = split_test
        train_label = split_train_label
        test_label = split_test_label
        
    else:
        train = pd.concat([train, split_train], ignore_index=True)
        test = pd.concat([test, split_test], ignore_index=True)
        
        train_label = pd.concat([train_label, split_train_label], ignore_index=True)
        test_label = pd.concat([test_label, split_test_label], ignore_index=True)
    
    
```

```python
#reseting all index information
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)

train_label = train_label.reset_index(drop=True)
test_label = test_label.reset_index(drop=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
for size in Knn_cluster_size.keys():
    knn_classifier_instance = KNeighborsClassifier(n_neighbors=size)
    
    accuracy_score_value = cross_val_score(knn_classifier_instance, train, 
                                   train_label, cv=5, scoring='accuracy')
    
    validation_error = (1 - accuracy_score_value.mean()) * 100
    Knn_cluster_size[size]['cross_validation_method']['validation_error'] = validation_error
    
    knn_classifier_instance.fit(train, train_label)
    prediction = knn_classifier_instance.predict(train)
    train_error = (1 - accuracy_score(train_label, prediction)) * 100
    Knn_cluster_size[size]['cross_validation_method']['training'] = train_error
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
stratified_obj = StratifiedKFold(n_splits=5)

for size in Knn_cluster_size.keys():
    knn_classifier_instance = KNeighborsClassifier(n_neighbors=size)
    
    for train_index, train_val_index in  stratified_obj.split(train,train_label):
        
        stratified_train=  pd.DataFrame(train, index=list(train_index))
        stratified_val=  pd.DataFrame(train, index=list(train_val_index))
    
        stratified_train_label = train_label.loc[list(train_index)]
        stratified_val_label = train_label.loc[list(train_val_index)]
        
        knn_classifier_instance.fit(stratified_train, stratified_train_label)
        
        predict= knn_classifier_instance.predict(stratified_val)
        validation_error= 1 - accuracy_score(stratified_val_label, predict)

        predict = knn_classifier_instance.predict(stratified_train)
        train_error = 1 - accuracy_score(stratified_train_label, predict)
        
        Knn_cluster_size[size]['stratified_cross_validation_method']['validation_error'] = validation_error*100
        Knn_cluster_size[size]['stratified_cross_validation_method']['training'] = train_error*100        
```

```python
#displaying the line chart
x = list(Knn_cluster_size.keys())
y_cross_validation = [entry['cross_validation_method']['validation_error'] for entry in Knn_cluster_size.values()]
y_cross_validation_training = [entry['cross_validation_method']['training'] for entry in Knn_cluster_size.values()]
y_stratified_cross_validation = [entry['stratified_cross_validation_method']['validation_error'] for entry in Knn_cluster_size.values()]
y_stratified_cross_validation_training = [entry['stratified_cross_validation_method']['training'] for entry in Knn_cluster_size.values()]

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(x, y_cross_validation, marker='o', label='Cross Validation Error')
plt.plot(x, y_cross_validation_training, marker='o', label='Cross Validation training error')
plt.plot(x, y_stratified_cross_validation, marker='o', label='Stratified Cross Validation Error')
plt.plot(x, y_stratified_cross_validation_training, marker='o', label='Stratified Cross Validation training error')

plt.title('KNN')
plt.xlabel('K-value')
plt.ylabel('mean validation error / training Error (%)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning) 
import numpy as np
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
def load_classification(classifier, train,train_label,test,test_label):
    prediction_label =''
    acutal=''
    accuracy =[]
    confusion_list =[]
    stratified_obj = StratifiedKFold(n_splits=5)
    
    for train_index, train_val_index in  stratified_obj.split(train,train_label):
        
        stratified_train=  pd.DataFrame(train, index=list(train_index))
        stratified_val=  pd.DataFrame(train, index=list(train_val_index))
    
        stratified_train_label = train_label.loc[list(train_index)]
        stratified_val_label = train_label.loc[list(train_val_index)]
        
        classifier.fit(stratified_train, stratified_train_label)
        
        predict= classifier.predict(stratified_val)
        validation_accuracy= accuracy_score(stratified_val_label, predict)
        accuracy.append(validation_accuracy)
        confusion_list.append(confusion_matrix(stratified_val_label, predict))
        
    classifier.fit(train, train_label)
    predict = classifier.predict(test)
    mode_accuracy = accuracy_score(test_label, predict)
    
    return accuracy,mode_accuracy,predict,test_label,confusion_list
        
```

```python
#MLP classifier
cross_validation_accuracy,accuracy,prediction_label,acutal_label,confusion_list= load_classification(MLPClassifier(hidden_layer_sizes=(10, 10, 10)), train,train_label,test,test_label)
```

```python
print("MLP classifier cross validation accuracy is {}".format(str(np.mean(cross_validation_accuracy))))
print("MLP classifier model Accuracy is {}".format(str(accuracy)))
f1_score_value = f1_score(acutal_label, prediction_label, average='weighted') 
print("MLP classifier F-1 score is "+str(f1_score_value))
confusion_matix_obj = np.mean(confusion_list,axis=0)
ConfusionMatrixDisplay(confusion_matrix=confusion_matix_obj,display_labels=config_dictionary['bread_info'].keys()).plot();
```
